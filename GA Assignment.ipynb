### If you could collect data on another attribute of these projects, what would it be and why? *

If there were any possible way to obtain the amount of money spent on marketing efforts, that would be my first choice attribute to collect data on.
Marketing efforts will bring awareness of a project on multiple platforms or channels, that would then convert an advertisement into a website visit, potentially converting a visit to a successful backing.
The importance of being able to collect information about each project's marketing efforts in a quantifiable way (i.e. dollar amount spent on marketing) would be a useful feature in analyzing the success and failure of kickstarter projects.

The second choice of attribute to collect data on would be the website views.

With the website views, we should be able to, with a relatively high confidence, use it as a factor to determine the rate of success as there is a "rate of converting" website views to project backing.

In terms of accessing useful data, we will drop all `status = live` projects as it is too time consuming to verify if it is successful or unsuccessful.

There are two other `statuses` which we will want to add into our "ignore list": they are `canceled` and `suspended`.
A quick research into kickstarter shows that <b>canceled</b> is typically triggered from the project creator, and the <b>suspended</b> happens when the project creator has violated some of kickstarter's rules and regulations. Since this is an external issue which is not in scope of this analysis, we will also remove it.

#### Fix the Film &amp; Video data, combine that with the `Film & Video` data

```python
df_clean['category_clean'] = df_clean['category'].str.replace('Film &amp; Video', 'Film & Video')
```

```python
df_clean['subcategory'].value_counts()
```

```python
def clean_subcategory(data):
    data['subcategory_clean'] = data['subcategory'].copy()
    data['subcategory_clean'] = data['subcategory_clean'].str.replace('Film &amp; Video', 'Film & Video')
    data['subcategory_clean'] = data['subcategory_clean'].str.replace('Country &amp; Folk', 'Country & Folk')
    data['subcategory_clean'] = data['subcategory_clean'].str.replace('Board &amp; Card Games', 'Board & Card Games')
    
    return data

df_clean = clean_subcategory(df_clean)
```

```python
# Segregate numerical and categorical columns
num_cols = ['name','status','goal','pledged','funded percentage','backers','levels','updates','comments','duration']
cat_cols = ['name','status','category_clean','subcategory_clean']
df_clean_numerical = df_clean[num_cols]
df_clean_categorical = df_clean[cat_cols]
```

- <i>Background information about pledge is that, typically a goal does not need to be high, as the project success rate is determined by whether the amount pledged can reach the goal in the set duration. ### Date

```python
df_clean.columns
```

```python
df_clean['funded date dt'] = df_clean['funded date'].apply(lambda x: x[5:16])
df_clean['funded date dt'] = pd.to_datetime(df_clean['funded date dt'])
df_clean['funded date dt']
```

```python
df_clean['launch date dt'] = (df_clean['funded date dt'] - pd.to_timedelta(df_clean['duration'].apply(lambda x: str(x) + 'D')))
```

```python
df_clean['launch date dt'] = pd.to_datetime(df_clean['launch date dt'])
df_clean['launch date dt']
```

```python
df_clean['launch year'] = df_clean['launch date dt'].dt.year
df_clean['launch month'] = df_clean['launch date dt'].dt.month
df_clean['funded year'] = df_clean['funded date dt'].dt.year
df_clean['funded month'] = df_clean['funded date dt'].dt.month
df_clean
```

```python
df_dates = df_clean[['status','launch year', 'launch month', 'funded year','funded month']]
# df_dates['launch yearmon'] = df_dates['launch year'].astype('string') + '-' + df_dates['launch month'].astype('string')
# df_dates['funded yearmon'] = df_dates['funded year'] + '-' + df_dates['funded month']
df_dates['launch yearmon'] = df_dates['launch year'].astype('str').str.cat(df_dates['launch month'].astype('str'), sep='-')
# df_dates['launch yearmon']
df_dates['funded yearmon'] = df_dates['funded year'].astype('str').str.cat(df_dates['funded month'].astype('str'), sep='-')
df_dates['funded yearmon']
```
